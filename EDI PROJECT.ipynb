```python
def hashtag_extract(text_corpus):
  hashtag=[]
  for text in text_corpus:
    ht=re.findall(r"#(\w+)",text)
    hashtag.append(ht)

  return hashtag
```

```python
def hashtag_freq(hashtag):
  a=nltk.FreqDist(hashtag)
  d=pd.DataFrame({'Hashtah':list(a.keys()),
                 'Freq': list(a.values())}
                 )
  d=d.nlargest(columns="Freq",n=30)

  return d
```

```python
def clean(text):
  text=str(text).lower()
  text=re.sub('https?://\S+|www\.\S+','',text)
  text=re.sub('[%s]'%re.escape(string.punctuation),'',text)
  text=re.sub('\n|\t','',text)
  text=re.sub('\d','',text)
  text=[word for word in text.split(' ') if word not in stopword]
  text=" ".join(text)
  text=[stemmer.stem(word) for word in text.split(' ')]
  text=" ".join(text)

  return text
```

```python
def get_sentiment(text):
    blob=TextBlob(text)
    sentiment=blob.sentiment.polarity
    if sentiment > 0:
        result ="Positive"
    elif sentiment < 0:
        result ="Negative"
    else:
        result="Neutral"
    return result
```

```python
from collections import Counter
```

```python
def extract_keywords(text,num=50):
    tokens = [tok for tok in text.split()]
    most_common_tokens=Counter(tokens).most_common(num)
    return dict(most_common_tokens)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical

import numpy as np

# Prepare the data
texts =df['Clean_Text']
labels =df['Emotion']

# Perform text tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)

# Pad sequences to have the same length
max_length = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_length)

# Convert labels to categorical
num_classes = len(set(labels))

# Convert labels to integers
label_mapping = {label: i for i, label in enumerate(set(labels))}
integer_labels = [label_mapping[label] for label in labels]

# Perform text tokenization, padding, and splitting as before

# Convert labels to categorical
num_classes = len(label_mapping)
categorical_labels = to_categorical(integer_labels, num_classes=num_classes)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, categorical_labels, test_size=0.2)

# Build the BiLSTM model
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=100, input_length=max_length))
model.add(Bidirectional(LSTM(64, return_sequences=True)))
model.add(Bidirectional(LSTM(64)))
model.add(Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, batch_size=32, epochs=7, validation_data=(X_test, y_test))

# Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)

# Make predictions
```

```python
def preprocess_text(input_sentence):
  input_sentence = pd.Series([input_sentence])
  input_sentence=input_sentence.apply(nfx.remove_stopwords)
  input_sentence=input_sentence.apply(nfx.remove_userhandles)
  input_sentence=input_sentence.apply(nfx.remove_punctuations)
  return input_sentence[0]
```

```python

# Preprocess the input sentence
def find_emo(input_sentence):
  preprocessed_sentence = preprocess_text(input_sentence)  # Preprocess the sentence as per your preprocessing steps

  # Convert the preprocessed sentence to numerical representation
  numerical_sentence = tokenizer.texts_to_sequences([preprocessed_sentence])
  padded_sentence = pad_sequences(numerical_sentence, maxlen=max_length)

  # Make the prediction
  predictions = model.predict(padded_sentence)
  predicted_class = np.argmax(predictions[0])
  list_val=[i for i in label_mapping]
  return list_val[predicted_class]

```
